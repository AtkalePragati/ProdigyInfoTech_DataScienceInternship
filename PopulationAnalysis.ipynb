```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dataframe = pd.read_csv('POPULATION.csv')
dataframe
```

```python

df_final = dataframe.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
chosen_country = 'India'


country_data = df_final[df_final['Country Name'] == chosen_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(8, 4))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')


plt.show()
```

```python
plt.hist(population, bins=40, color='skyblue', edgecolor='darkblue')
plt.show()
```
